# How to handle time series data with ease?
Using pandas datetime properties# 

```python
df_no2=pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/air_quality_no2_long.csv")
df_no2.head()
```

```python
# rename column name
df_no2=df_no2.rename(columns={"date.utc":"datetime"})
df_no2
```

```python
df_no2.city.unique()
```

```python
df_no2.dtypes
```

```python
df_no2["datetime"]=pd.to_datetime(df_no2["datetime"])
df_no2.dtypes
```

I want to add a new column to the DataFrame containing only the MONTH of the measurement

```python
df_no2["month"]=df_no2["datetime"].dt.month
df_no2.head()
```

I want to add a new column to the DataFrame containing only the DATE of the measurement

```python
df_no2["date"]=df_no2["datetime"].dt.date
df_no2.head()
```

```python
df_no2["year"]=df_no2["datetime"].dt.year
df_no2.head()
```

```python
import matplotlib.pyplot as plt
df_no2.groupby(df_no2["datetime"].dt.hour)["value"].mean().plot(figsize=(10,10),kind='bar', rot=0)
plt.xlabel("Hour of the day") # custom x label using Matplotlib

plt.ylabel("$NO_2 (µg/m^3)$")
```
